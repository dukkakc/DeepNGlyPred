```python
import re
from Bio import SeqIO

all_motPos_LS=[]
motif_pos_LS=[]
neighborSeq_LS=[]
incompleteSeq_LS=[]
missingSeq_LS=[]

reqd_ws = 51
 
for seq_record in SeqIO.parse("3080_rework_Singlefile.fasta","fasta"):
    placeholder = seq_record.id  
    seq = str(seq_record.seq)
    placeholder = placeholder[0:6]
       
    for i, j in enumerate(seq):

            if j in "N" and ((i+3)<len(seq)):

                if ((seq[i+1] != 'P') and (seq[i+2] == 'S'or seq[i+2] == 'T')):        

                    mot_pos=(i+1)
                    all_motPos_LS.append(mot_pos)                                          

                    mot_pat=seq[i:(i+3)]                                                    
                    seqWin=seq[i-25:i+26].replace('\n', '').replace('\r', '')            

                    if len(seqWin) == 51:
                        filenamename = placeholder
                        with open("Oct_21_AA.txt","a+") as infile:
                            infile.write(">")
                            infile.write(filenamename)
                            infile.write("   ")
                            infile.write("|")
                            infile.write("   ")
                            infile.write(str(mot_pos))
                            infile.write("   ")
                            infile.write(seqWin)
                            infile.write("\n")




                    else:
                        incompleteSeq_LS.append(mot_pos)

                        fill=[]

                        for x in range(reqd_ws):
                            dummy="-"
                            fill.append(dummy)


                        if int(mot_pos)<26:                
                            avail=reqd_ws-int(mot_pos)


                            nSlice=seq[i-(mot_pos-1): i+26]    

                            partial=len(nSlice)

                            avail_partial=int(reqd_ws)-int(partial)  


                            select=fill[0:int(avail_partial)]      
                            select.extend(nSlice)                  

                            nTer_adjust=''.join(select)
                            filenamename = placeholder
                           
                            with open("Oct_21_AA.txt","a+") as infile:
                                infile.write(">")
                                infile.write(filenamename)
                                infile.write("   ")
                                infile.write("|")
                                infile.write("   ")
                                infile.write(str(mot_pos))
                                infile.write("   ")
                                infile.write(nTer_adjust)
                                infile.write("\n")


                            missingSeq_LS.append(nTer_adjust)

   


                        else:


                            if int(mot_pos)>26:

                                available=int(len(seq)-int(mot_pos))                

                                cSlice=seq[i-25:i+(int(available)+1)].replace('\n', '').replace('\r', '')      


                                cLength=len(cSlice)

                                avail_cLength= int(reqd_ws)-int(cLength)        


                                select=fill[0:int(avail_cLength)]              

                                imperfect=''.join(select)                      

                                cTer_adjust=cSlice+imperfect  
                                filenamename = placeholder
                                with open("Oct_21.txt","a+") as infile:
                                    infile.write(">")
                                    infile.write(filenamename)
                                    infile.write("   ")
                                    infile.write("|")
                                    infile.write("   ")
                                    infile.write(str(mot_pos))
                                    infile.write("   ")
                                    infile.write(cTer_adjust)
                                    infile.write("\n")


                                missingSeq_LS.append(cTer_adjust)              


                                miss_Cter_sequenceWindow_DI=dict(zip(incompleteSeq_LS, missingSeq_LS))           
```
